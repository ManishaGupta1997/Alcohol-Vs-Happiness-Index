### Last approach:- 
After fitting all those models, our last variable which was region is left. Now we will do one-hot encoding and then add it as one of the variable. 

#### One hot encoding

```python
df2 = pd.read_csv(r"C:\Users\manis\Desktop\HappinessAlcoholConsumption.csv")
df2.head()
```

```python
df2 = df2.drop(['Country', 'Hemisphere'],axis=1)
```

```python
#creating a new column total_alcohol
df2['total_alcohol']=df2['Beer_PerCapita']+ df2['Spirit_PerCapita']+ df2['Wine_PerCapita']
```

```python
# Taking only numerical variables for standardization
stand = df2[['HappinessScore','HDI','GDP_PerCapita','Beer_PerCapita','Spirit_PerCapita','Wine_PerCapita','total_alcohol']]
```

```python
from sklearn.preprocessing import StandardScaler
# define standard scaler
scaler = StandardScaler()
# transform data
scaled = scaler.fit_transform(stand)
print(scaled)
```

```python
index = stand.index
col = stand.columns
```

```python
stand = pd.DataFrame(scaled, index=index, columns=col)
stand.head()
```

```python
print(df2['Region'].unique()) #printing unique regions
```

```python
df2['Region'].value_counts()  #Checking for the label counts in the categorical parameters
```

```python
#One-Hot encoding the categorical parameters using get_dummies()
dummies = pd.get_dummies(df2['Region'])
dummies
```

### Concatenating dummies to standardized values 

```python
merged = pd.concat([stand,dummies],axis=1)
merged
```

```python
x1 = merged.drop("HappinessScore",axis=1)
y1 = merged['HappinessScore']

#train and test data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x1,y1,test_size=0.3,random_state=45)


#linear regression
from sklearn.linear_model import LinearRegression
linear = LinearRegression()
linear.fit(x_train,y_train)
print('Linear Regression= ',linear.score(x_test,y_test))

#decision tree
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state = 45)
regressor.fit(x_train, y_train)
print('Decision Tree= ',regressor.score(x_test,y_test))

#random forest
from sklearn.ensemble import RandomForestRegressor
regressor_forest = RandomForestRegressor(n_estimators = 10, random_state = 42)
regressor_forest.fit(x_train, y_train)
print('Random Forest= ',regressor_forest.score(x_test,y_test))

#Gradient Boosting
from sklearn.ensemble import GradientBoostingRegressor
boost = GradientBoostingRegressor()
boost.fit(x_train,y_train)
print('Gradient Boosting= ',boost.score(x_test,y_test))

#Ridge
from sklearn.linear_model import Ridge
ridge = Ridge(alpha=1)
ridge.fit(x_train,y_train)
print('Ridge= ',ridge.score(x_test,y_test))

#AdaBoost
from sklearn.ensemble import AdaBoostRegressor
ada = AdaBoostRegressor()
ada.fit(x_train,y_train)
print('AdaBoost= ',ada.score(x_test,y_test))
```

#### Observation:
Gradient boosting seems to perform better than any other model here. 

This made us think that Region also played a major role in identifying happiness score. 

## FINAL OBSERVATION:-
1. Interestingly, despite what was our hypothesis, it was found that there is no observable relationship between alcohol consumption and happiness. This was confirmed by using regression analysis on plots of the Happiness Score versus alcohol consumption (for wine, beer and spirits). 

2. From the initial data analysis a range of plots was made to get an overview of any possible linear relationships. From this it appeared that GDP was related to happiness in some way. By using regression analysis, it was found that GDP had an exponential relationship with Happiness Score. Also, HDI showed linear relationship with happiness score. 

3. As the GDP increases, the initial increase in the Happiness Score is large. For the country Syria, it has a low GDP and this results in a low Happiness Score due to the poor economic conditions of the country. This allows us to conclude that there exists a certain threshold value for GDP for which it no longer causes an increase in the Happiness Score. When this is reached, the Happiness Score could potentially be influenced by other factors. This means that alcohol consumption may have a relationship with the Happiness Score for countries with higher GDP values.
4. As the HDI increases, there is an increase in Happiness score as well. As an Indian proverb says, it is better to be blind than to see things only from one point of view. And so it is with looking at peopleâ€™s development through only the happiness or alcohol consumption. Combining major other factors can provide a richer and more accurate picture of what is happening.
